### In this notebook we will look at the developement of wordcloud

```python
from wordcloud import WordCloud
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from pylab import rcParams
rcParams['figure.figsize'] = 6,6
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
```

```python
lot = ""
for i in range(len(df)):
    lot += "".join(df.iloc[i]['v2'])
```

```python
#text = df['v2']

# Create and generate a word cloud image:
wordcloud = WordCloud(width = 600, height = 600, background_color="black").generate(lot)

# Display the generated image:
plt.figure(figsize = (8, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(width = 600, height = 600, background_color="white").generate(lot)
plt.figure(figsize = (6, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"Youtube04-Eminem.csv", encoding ="latin-1")
  
comment_words = ''
stopwords = set(STOPWORDS)
  
# iterate through the csv file
for val in df.CONTENT:
      
    # typecaste each val to string
    val = str(val)
  
    # split the value
    tokens = val.split()
      
    # Converts each token into lowercase
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 600, height = 600,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
  
# plot the WordCloud image                       
plt.figure(figsize = (6, 6), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```

```python
wine_mask = np.array(Image.open("wine_mask_xgk1tq.png"))
#wine_mask = wine_mask[:,:,1]
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]), np.int32)

for i in range(len(wine_mask)):
    transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
```

```python
wc = WordCloud(background_color="white", max_words=1000, mask=transformed_wine_mask,
               stopwords=stopwords, contour_width=3, contour_color='firebrick')

# Generate a wordcloud
wc.generate(lot)

# show
plt.figure(figsize=[20,10])
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
